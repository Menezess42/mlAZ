```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_csv('./data/Mall_Customers.csv')
x = dataset.iloc[:, [3, 4]].values # keeping just two columns for visualysing purposes
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init = 'k-means++', random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(x)
print(y_kmeans)
```

```python
plt.scatter(x[y_kmeans==0,0],
            x[y_kmeans==0,1],
            s = 100,
            c = 'red',
            label = 'cluster 1') # each x call is a cluster

plt.scatter(x[y_kmeans==1,0],
            x[y_kmeans==1,1],
            s = 100,
            c = 'blue',
            label = 'cluster 2') # each x call is a cluster

plt.scatter(x[y_kmeans==2,0],
            x[y_kmeans==2,1],
            s = 100,
            c = 'green',
            label = 'cluster 3') # each x call is a cluster

plt.scatter(x[y_kmeans==3,0],
            x[y_kmeans==3,1],
            s = 100,
            c = 'cyan',
            label = 'cluster 4') # each x call is a cluster

plt.scatter(x[y_kmeans==4,0],
            x[y_kmeans==4,1],
            s = 100,
            c = 'magenta',
            label = 'cluster 5') # each x call is a cluster

plt.scatter(kmeans.cluster_centers_[:, 0],
            kmeans.cluster_centers_[:, 1],
            s = 300,
            c = 'yellow',
            label = 'centeroids')

plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
