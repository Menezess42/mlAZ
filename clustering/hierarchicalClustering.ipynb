```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dataset = pd.read_csv('./data/Mall_Customers.csv')
x = dataset.iloc[:, [3, 4]].values # keeping just two columns for visualysing purposes
```

```python
linkage = sch.linkage(x, method='ward')
dendrogram = sch.dendrogram(linkage)

plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

```python
hc = AgglomerativeClustering(n_clusters = 5,
                             metric = 'euclidean',
                             linkage = 'ward')
y_hc = hc.fit_predict(x)
```

```python
print(y_hc)
```

```python
plt.scatter(x[y_hc==0,0],
            x[y_hc==0,1],
            s = 100,
            c = 'red',
            label = 'cluster 1') # each x call is a cluster

plt.scatter(x[y_hc==1,0],
            x[y_hc==1,1],
            s = 100,
            c = 'blue',
            label = 'cluster 2') # each x call is a cluster

plt.scatter(x[y_hc==2,0],
            x[y_hc==2,1],
            s = 100,
            c = 'green',
            label = 'cluster 3') # each x call is a cluster

plt.scatter(x[y_hc==3,0],
            x[y_hc==3,1],
            s = 100,
            c = 'cyan',
            label = 'cluster 4') # each x call is a cluster

plt.scatter(x[y_hc==4,0],
            x[y_hc==4,1],
            s = 100,
            c = 'magenta',
            label = 'cluster 5') # each x call is a cluster

plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
