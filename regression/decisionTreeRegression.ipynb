```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('./data/Position_Salaries.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.tree import DecisionTreeRegressor

dTree_regressor = DecisionTreeRegressor(random_state=0)

dTree_regressor.fit(x, y)
```

```python
y_pred = dTree_regressor.predict([[6.5]])
```

This tool do not work wel with dataset of 1 feature like
the one I am using here (position_salaries.csv)
I just want to test and see how dTree works

```python
x_grid = np.arange(x.min(), x.max(), 0.1)
x_grid = x_grid.reshape(len(x_grid), 1)

plt.scatter(x, y, color='red')
plt.plot(x_grid, dTree_regressor.predict(x_grid), color = 'blue')
plt.title('True or Bluff (Decision Tree Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
