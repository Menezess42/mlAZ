```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('./data/Position_Salaries.csv')
x = dataset.iloc[:, 1 :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
```

```python
print(y)
```

```python
y = y.reshape(len(y), 1)
print(y)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

standardScaler_x = StandardScaler()
standardScaler_y = StandardScaler()

x = standardScaler_x.fit_transform(x)
y = standardScaler_y.fit_transform(y)
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.svm import SVR

svr_regressor = SVR(kernel = 'rbf')

svr_regressor.fit(x, y)
```

```python
prediction = svr_regressor.predict(standardScaler_x.transform([[6.5]]))
print(prediction)

reversed_y = standardScaler_y.inverse_transform(prediction.reshape(-1,1))
print(reversed_y)
```

```python
reversed_x = standardScaler_x.inverse_transform(x)
reversed_y = standardScaler_y.inverse_transform(y)


y_pred = svr_regressor.predict(x)
reversed_y_pred = standardScaler_y.inverse_transform(y_pred.reshape(-1,1))

plt.scatter(reversed_x, reversed_y, color='red')
plt.plot(reversed_x, reversed_y_pred, color = 'blue')
plt.title('truth or Bluff (SVR)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
reversed_x = standardScaler_x.inverse_transform(x) # ok ok

x_grid = np.arange(reversed_x.min(), reversed_x.max(), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))

reversed_y = standardScaler_y.inverse_transform(y) # ok

scaled_x_grid = standardScaler_x.transform(x_grid)
y_pred = svr_regressor.predict(scaled_x_grid)
reversed_y_pred = standardScaler_y.inverse_transform(y_pred.reshape(-1,1))

plt.scatter(reversed_x, reversed_y, color='red')
plt.plot(x_grid, reversed_y_pred, color = 'blue')
plt.title('truth or Bluff (SVR)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
