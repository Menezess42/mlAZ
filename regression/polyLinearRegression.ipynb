```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('./data/Position_Salaries.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
# Simple Linear Regression (just remebering. Not useful for this code)
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(x, y)
```

```python
# Poly Linear Regression
from sklearn.preprocessing import PolynomialFeatures
poly_linear_regression = PolynomialFeatures(degree=2) # if we increase the degree to 4, we will have overfiting
x_polynomial = poly_linear_regression.fit_transform(x)

# We aply the x_polynomial as input into a linear regressor
linear_regressor_2 = LinearRegression()
linear_regressor_2.fit(x_polynomial, y)
```

```python
# Visualising the linear regession results
plt.scatter(x, y, color='red')
plt.plot(x, linear_regressor.predict(x), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

Reds dots are the real salary.
Blue line is the predicted values for a simple linear regression.

We can see here that simple linear regression is not useful in this case

```python
# Visualising the poly linear regession results
plt.scatter(x, y, color='red')
plt.plot(x, linear_regressor_2.predict(x_polynomial), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Predicting a new result with linear regression
linear_regressor.predict([[6.5]])

# Linear regression predicts to over the baseline to a position level o 6.5
```

```python
# Predicting a new result with polynomial linear regression
linear_regressor_2.predict(poly_linear_regression.fit_transform([[6.5]]))

# Poly linear regression predicts ok base on the baseline to a position of 6.5
```
